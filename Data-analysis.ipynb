```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read train.csv file
train = np.genfromtxt('COMP90086_2024_Project_train/train.csv', delimiter=',', skip_header=1)
print(train.shape)
id = train[:, 0] 
shapeset = train[:, 1] #1: only cubes, 2: multiple shapes
type = train[:, 2] #1: easy, 2: hard
total_height = train[:, 3] # number of objects
instability_type = train[:, 4] #0: no instability, 1: centre of mass, 2: non-planar surface
cam_angle = train[:, 5] # 1: low, 2: high
stable_height = train[:, 6]
```

```python
# Plot the type
plt.hist(type, bins=np.arange(1, 4) - 0.5, rwidth=0.3)
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.title('Type')
plt.xticks([1, 2])  # Setting x-ticks to match the values
plt.show()
```

```python
# Plot the shapeset
plt.hist(shapeset, bins=np.arange(1, 4) - 0.5, rwidth=0.3)
plt.xlabel('Shapeset')
plt.ylabel('Frequency')
plt.title('Shapeset')
plt.xticks([1, 2])  # Setting x-ticks to match the values
plt.show()
```

```python
# Plot the total_height
plt.hist(total_height, bins=np.arange(1, 11) - 0.5, rwidth=0.3)
plt.xlabel('Total Height')
plt.ylabel('Frequency')
plt.title('Total Height')
plt.xticks(np.arange(1, 11))  # Setting x-ticks to match the values
plt.show()
```

```python
# Plot the instability_type
plt.hist(instability_type, bins=np.arange(0, 4) - 0.5, rwidth=0.3)
plt.xlabel('Instability Type')
plt.ylabel('Frequency')
plt.title('Instability Type')
plt.xticks([0, 1, 2])  # Setting x-ticks to match the values
plt.show()
```

```python
# Plot the cam_angle
plt.hist(cam_angle, bins=np.arange(1, 4) - 0.5, rwidth=0.3)
plt.xlabel('Camera Angle')
plt.ylabel('Frequency')
plt.title('Camera Angle')
plt.xticks([1, 2])  # Setting x-ticks to match the values
plt.show()
```

```python
# Plot the stable_height
plt.hist(stable_height, bins=np.arange(1, 11) - 0.5, rwidth=0.3)
plt.xlabel('Stable Height')
plt.ylabel('Frequency')
plt.title('Stable Height')
plt.xticks(np.arange(1, 11))  # Setting x-ticks to match the values
plt.show()
```

```python
# Plot the past 6 figures together in one figure
fig, axs = plt.subplots(2, 3)
fig.suptitle('All Features')
axs[0, 0].hist(type, bins=np.arange(1, 4) - 0.5, rwidth=0.3)
axs[0, 0].set_title('Type')
axs[0, 0].set_xticks([1, 2])
axs[0, 1].hist(shapeset, bins=np.arange(1, 4) - 0.5, rwidth=0.3)
axs[0, 1].set_title('Shapeset')
axs[0, 1].set_xticks([1, 2])
axs[0, 2].hist(total_height, bins=np.arange(1, 11) - 0.5, rwidth=0.3)
axs[0, 2].set_title('Total Height')
axs[0, 2].set_xticks(np.arange(1, 11))
axs[1, 0].hist(instability_type, bins=np.arange(0, 4) - 0.5, rwidth=0.3)
axs[1, 0].set_title('Instability Type')
axs[1, 0].set_xticks([0, 1, 2])
axs[1, 1].hist(cam_angle, bins=np.arange(1, 4) - 0.5, rwidth=0.3)
axs[1, 1].set_title('Camera Angle')
axs[1, 1].set_xticks([1, 2])
axs[1, 2].hist(stable_height, bins=np.arange(1, 11) - 0.5, rwidth=0.3)
axs[1, 2].set_title('Stable Height')
axs[1, 2].set_xticks(np.arange(1, 11))
plt.show()
```

```python
def show_train_data(id_):
    # index of id_ in the id array
    index = np.where(id == id_)[0][0]
    img = plt.imread(f'COMP90086_2024_Project_train/train/{int(id_)}.jpg')
    plt.axis('off')
    plt.imshow(img)
    type_ = 'only has cubes' if int(type[index]) == 1 else 'has multiple shape types'
    reason = 'no stability issues'
    if int(instability_type[index]) == 1:
        reason = 'an unsupported centre of mass'
    elif int(instability_type[index]) == 2:
        reason = 'a non-planar surface'
    cam_angle_ = 'low' if int(cam_angle[index]) == 1 else 'high'
    difficulty = 'easy' if int(type[index]) == 1 else 'hard'
    print(f'Image {int(id_)} is stable until {int(stable_height[index])} blocks because it has {reason}. It {type_} and a total height of {int(total_height[index])}. Camera angle is {cam_angle_} and the task is {difficulty}.')
```

```python
show_train_data(id[0])
```

```python
show_train_data(id[1])
```

```python
show_train_data(id[3])
```
